### Part 1: The Dataset



```python
#Step 1: Loading the Dataset From a File.

# Code cell 1
import pandas as pd
brainFile = 'brainsize.txt'
brainFrame = pd.read_csv(brainFile, '\t')

#Before we do anything else, first, we have to upload the dataset into the memory.
```

```python
#Step 2: Verifying the dataframe
brainFrame.head()
#To display the header of the dataset we use the "head()" to show the first 5 entries of the dataset.
```

### Part 2: Scatterplot Graphs and Correlatable Variables

```python
#Step 1: The pandas describe() method.

# Code cell 3
brainFrame.describe()

#We use the "describe()" to show the common values a dataset have.
```

```python
#Step 2: Scatterplot graphs

#a. Load the required modules.
import numpy as np
import matplotlib.pyplot as plt

#b. Separate the data.
#men
menDF = brainFrame[(brainFrame.Gender == 'Male')]

#c. Plot the graphs
menMeanSmarts = menDF[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDF["MRI_Count"])
plt.show()
%matplotlib inline

#this snippet of code shows the scatterplot graph of all the men entries
```

```python
#women
womenDF = brainFrame[(brainFrame.Gender == 'Female')]
womenMeanSmarts = womenDF[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDF["MRI_Count"])
plt.show()
%matplotlib inline
#this snippet shows the scatterplot graph of women entries
```

### Part 3: Calculating Correlation with Python

```python
#Step 1: Calculate correlation against brainFrame.

brainFrame.corr(method='pearson')
```

##Question
Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filled
with 1s? Is that a coincidence? Explain. <br>
- It is not coincidence, it is diagonal filled with 1s because we used pearson correlation which measures the linear relationshiop between 2 variables and the correlation of each variable to its self is always 1.

Still looking at the correlation table above, notice that the values are mirrored; values below the 1
diagonal have a mirrored counterpart above the 1 diagonal. Is that a coincidence? Explain. <br>
- No, it is also not a coincidence, when you look at the variables in the table, they are all closely related to each other which shows that the  correlation between these variables even when you jumble it, it will stay the same.

```python
womenDF.corr(method='pearson')
```

```python
menDF.corr(method='pearson')
```
